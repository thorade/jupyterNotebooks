```python
import urllib.request
import zipfile

url = "http://projects.buildingsmartalliance.org/files/?artifact_id=4278"
outname = "duplex.zip"

urllib.request.urlretrieve(url, outname)

with zipfile.ZipFile(outname,"r") as zip_ref:
    zip_ref.extractall()
```

```python
import ifcopenshell
ifc_file = ifcopenshell.open("Duplex_A_20110907_optimized.ifc")
# filter by type
walls = ifc_file.by_type("IfcWall")
for wall in walls:
    print("\nnext wall: ", wall.id)
```
